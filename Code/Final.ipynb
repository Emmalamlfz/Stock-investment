```python
import os
import glob
import pandas as pd

# Get the current file path
folder_path = os.getcwd()
# Read the name of all excel files
excel_files = glob.glob(os.path.join(folder_path, "*.xlsx"))
```

```python
# 读取当前文件目录中所有的excel文件名
folder_path = os.getcwd()
excel_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx')]

# 读取excel文件为dataframe并删除带有空值的行
dataframes = {}
for file in excel_files:
    # 读取excel文件为dataframe
    print(file)
    df = pd.read_excel(os.path.join(folder_path, file), engine='openpyxl')
    # 删除带有空值的行
    df.dropna(subset=['收盘价(元)'], inplace=True)
    df = df[df['日期'] > '2022-01-23']
    df = df.reset_index(drop = True)
    df['Returen_daily'] = df['收盘价(元)'] / df['开盘价(元)'] - 1
    # 将dataframe添加到字典中，字典的键为文件名的去除'.xlsx'后缀
    dataframes[file[:-5]] = df
```

```python
# Define function to calculate the market return for one day
def get_market_return(df):
    #Get the Weighted Average Return for the market profolio
    total_market_cap = df['总市值2'].sum()
    df['weight'] = df['总市值2'] / total_market_cap
    weighted_return = (df['Returen_daily'] * df['weight']).sum()
    return weighted_return
```

```python
# Define function to calculate the factor return for SMB one day:
def get_SMB_return(df,n):
    df = df.sort_values(by = '总市值2',ascending = True)
    df = df.reset_index(drop = True)
    Top5 = df[:n]
    Bot5 = df[-n:]
    Top_Return = get_market_return(Top5)
    Bot_Return = get_market_return(Bot5)
    return (Top_Return - Bot_Return)
```

```python
# Define function to calculate the factor return for HMI one day:
def get_HMI_return(df,n):
    df['BM'] = df['总市值2'] / df['归属母公司股东的普通股权益(MRQ)']
    df = df.sort_values(by = 'BM',ascending = False)
    df = df.reset_index(drop = True)
    Top5 = df[:n]
    Bot5 = df[-n:]
    Top_Return = get_market_return(Top5)
    Bot_Return = get_market_return(Bot5)
    return (Top_Return - Bot_Return)
```

```python
from datetime import datetime
date_list = Whole_stock_dataset['日期'].unique().tolist()
date_l = []
for date in date_list:
# 将时间戳转换为datetime对象
    dt = datetime.fromtimestamp(int(date / 1e9))

# 将datetime对象格式化为日期字符串
    dt = dt.strftime('%Y-%m-%d')
    date_l.append(dt)
# 输出结果
```

```python
factor_list = pd.DataFrame()
for date in date_l:
    df = Whole_stock_dataset[Whole_stock_dataset['日期'] == date]
    market_return = get_market_return(df)
    SMB_return = get_SMB_return(df,5)
    HMI_return = get_HMI_return(df,5)
    return_info = {'Date':[date],'Market_Return':[market_return],'E(SMB)':[SMB_return],'E(HMI)':[HMI_return]}
    return_info = pd.DataFrame(return_info)
    factor_list = pd.concat([return_info,factor_list])
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np


def multi_factor_model(factor_list, stock_info, date, n):
    # 将日期字符串转换为 datetime 对象
    date = pd.to_datetime(date)
    # 回滚n天得到起始日期
    start_date = date - pd.DateOffset(n)
    
    # 根据起始日期和结束日期筛选出对应的 factor_list 数据
    factor_data = factor_list[(factor_list['Date'] >= start_date) & (factor_list['Date'] <= date)]
    # 根据起始日期和结束日期筛选出对应的 stock_info 数据
    stock_data = stock_info[(stock_info['日期'] >= start_date) & (stock_info['日期'] <= date)]
    
    # 将 factor_data 和 stock_data 进行合并
    data = pd.merge(factor_data, stock_data, left_on='Date',right_on = '日期', how = 'left')
    # 检查数据中是否包含缺失值或无穷大值
    if data.isnull().values.any() or data.isin([np.inf, -np.inf]).values.any():
    # 删除包含缺失值或无穷大值的行
        data = data.dropna(subset=['Rm-Rf'])
# 检查筛选后的数据是否至少包含两个数据点
    if len(data) < 2:
        return None

    # 拟合多因子模型并获取结果
    X = data[['E(SMB)', 'E(HMI)', 'Rm-Rf']]
    y = data['Returen_daily'] - data['Rf']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    
    # 获取 alpha 值
    alpha = results.params['const']
    
    return alpha
```

```python
import datetime
# 设定回滚天数为60
n = 60

# 设置要遍历的日期范围：2023年1月4日至2023年3月31日
start_date = datetime.date(2023, 1, 3)
end_date = datetime.date(2023, 3, 31)
date_range = pd.date_range(start_date, end_date)
factor_list['Date'] = pd.to_datetime(factor_list['Date'])
# 初始化结果数据框
result_df = pd.DataFrame(columns=['Stock', 'Date', 'Alpha'])

# 遍历股票和日期
for stock in stock_list:
    stock_info = dataframes[stock]
    for date in date_range:
        alpha = multi_factor_model(factor_list, stock_info, date, n)
        result_df = result_df.append({'Stock': stock, 'Date': date, 'Alpha': alpha}, ignore_index=True)      
#         print(f"股票: {stock}, 日期: {date.strftime('%Y-%m-%d')}, alpha值: {alpha:.5f}")
output_folder = r"C:\Users\XIESITENG\Desktop\Final"  # 请将此路径更改为您想要保存Excel文件的文件夹
result_file_path = f"{output_folder}\\result_df.xlsx"
result_df.to_excel(result_file_path, index=False)
```
